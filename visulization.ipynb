```python
# importing the dataset
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Downloads/covid_data_kerala.csv")
df# Histogram
plt.figure(figsize=(10,6))
plt.hist(df['Confirmed'],bins=20,color='green',edgecolor='black')
plt.title('Histogram of COVID-19 Confirmed Cases in kerala')
plt.xlabel('Confirmed Cases')
plt.ylabel('Frequency')
plt.show()#confirmed and recovered case
plt.figure(figsize=(10,6))
plt.scatter(df['Confirmed'],df['Recovered'],color='blue',alpha=0.7)
plt.title('Scatter plot of COVID-19 Confirmed vs Recoved Cases in Kerala')
plt.xlabel('Confirmed Cases')
plt.ylabel('Recovered Cases')

#Adjust layout for better visualization
plt.tight_layout()
plt.show()
```

###### exception handling

```python
try:
    # code that might raise an exception
    result = 10/0  #will raise a zero division error
except ZeroDivisionError:
    # code to handle the specific exception
    print("Error: Cannot divide by zero.")
```

```python
try:
    #code
    num = int(input("Enter a number:"))
    result = 10 / num
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
except ValueError:
    print("Error: Invalid input. Please enter a valid number.")
else:
    #code to execute if no exception is raised
    print("Result:", result)
```

```python
try:
    #code
    num = int(input("Enter a number:"))
    result = 10 / num
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
except ValueError:
    print("Error: Invalid input. Please enter a valid number.")
except Exception as e:
    print("some error occured",e)
```

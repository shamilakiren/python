### MORE ON PANDAS PRE-PROCESSING

```python
import pandas as pd
import numpy as np
```

```python
d={
    'team':["SA","New zealand","Australia","India",
            "SA","New zealand","Australia","India",
            "New zealand","Australia","New zealand","Australia"],
    'Run':[330,230,300,180,200,250,190,400,340,290,390,333],
    "Match":[2,4,2,1,5,6,3,2,5,3,1,1],
    "Year":['2012','2012','2012','2012','2013','2013','2013','2013','2014','2014','2014','2014']
  }
df=pd.DataFrame(d)
df
```

```python
#GROUPBY
g=df.groupby('team')# returns an object pointing to its dataframe
g
```

```python
for team,teamdata in g:
    print(team)
    print("-"*10)
    print(teamdata)
    print("-"*35)
```

```python
g.get_group('SA') #RETURNS dataframes
```

```python
g.get_group('Australia')
```

```python
g.min()
```

```python
#g.mean()
```

```python
# Only apply mean to numeric columns
result = g.mean(numeric_only=True)
```

```python
#Aggregate

g_yr=df.groupby('Year')
g_yr['Run'].agg([np.min,np.max,np.mean,np.sum,np.std])
```

###concatenate rows

```python
NewYork={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['hot','sunny','rainy','cold'],
      "wind-speed":[33,23,45,24],
      "temp":[20,21,15,18]
}
NY=pd.DataFrame(NewYork)
NY
```

```python
Paris={
    "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
    "event":['sunny','cold','cold','rainy'],
    "wind-speed":[12,10,9,14],
    "temp":[14,16,15,10]
}
ps=pd.DataFrame(Paris)
ps
```

```python
df=pd.concat([NY,ps])
df
```

```python
df=pd.concat([NY,ps],ignore_index=True)
df
```

```python
df=pd.concat([NY,ps],keys=['New York data','Paris data'])
df
```

## concatenate columns

```python
temp1={
    "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
    "event":['sunny','cold','cold','rainy'],
    "temp":[14,16,15,10]
}
temp_a=pd.DataFrame(temp1)
temp_a
```

```python
temp2={
    "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
    "event":['sunny','cold','cold','rainy'],
    "wind-speed":[12,10,9,14],
}
temp_b=pd.DataFrame(temp2)
temp_b
```

```python
df=pd.concat([temp_a,temp_b],axis=1)
df
```

## join

```python
#inner join (returns only the matching rows)
df=pd.merge(temp_a,temp_b,on=['date','event'])
df
```

```python
df1=pd.merge(temp_a,temp_b,on=['date','event'],how='outer')
df1
```

```python
#outer join (Returns all rows)

temp_c={
    "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
    "event":['sunny','cold','hot','sunny'],
    "temp":[14,16,15,10]
}
temp=pd.DataFrame(temp_c)
temp
```

```python
temp_d={
    "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
    "event":['sunny','cold','hot','sunny'],
    "temp":[14,16,15,10]
}
wind_speed=pd.DataFrame(temp_d)
wind_speed
```

```python
df=pd.merge(temp,wind_speed,on=['date','event'],how='outer')
df
```

```python
#left join
df=pd.merge(temp,wind_speed,on=['date','event'],how='left')
df
```

```python
#right join

df=pd.merge(temp,wind_speed,on=['date','event'],how='right')
df
```
